```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def str_cleaner(s):
    k = s.replace('.','')
    k = k[:2] + '.' + k[2:]
    return float(k)
```

```python
df = pd.read_csv('data_75m.csv',  sep=', ', delimiter= ';')
df['Latitude'] = df['Latitude'].apply(str_cleaner)
df['Longitude'] = df['Longitude'].apply(str_cleaner)
df['Antena_Latitude'] = 55.472693
df['Antena_Longitude'] = 10.329273 
df['Antena_height'] = 22.5
```

```python
df
```

```python
df.columns
```

```python
cols = [ 'Latitude', 'Longitude', 'Altitude', 'SSS-RSRP', 'Antena_Latitude', 'Antena_Longitude',  'Antena_height', 'MAX_RSRP' ]
```

```python
custom_df = df[cols]
custom_df['Altitude'] = 75
```

```python
train_df = custom_df.iloc[:10000]
train_df
```

```python
train_df = custom_df.iloc[:10000]
test_df = custom_df.iloc[10000:]
train_ip = train_df[['Latitude', 'Longitude', 'Altitude', 'Antena_Latitude', 'Antena_Longitude', 'Antena_height' ]]
train_op = train_df['SSS-RSRP']

test_ip = test_df[['Latitude', 'Longitude', 'Altitude', 'Antena_Latitude', 'Antena_Longitude', 'Antena_height'  ]]
test_op = test_df['SSS-RSRP']
```

```python
train_op
```

```python
reg = LinearRegression().fit(train_ip, train_op)
```

```python
pred_test_op = reg.predict(test_ip)
```

```python
## Error Ca;culation
error = mean_squared_error(test_op, pred_test_op)
error
```
